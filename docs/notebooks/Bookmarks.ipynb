# Bookmark Usage

 This notebook demonstrates how to create, save and reuse bookmarks


## Script points the map to Mt Etna
Adjust your image as desired, the create a bookmark named Mt Etna

Script also demonstrates the use of the navigation controls (+/- Zoom, and North tools)

```python
 %pip install anymap-ts
```

```python
from anymap_ts import Map
from pathlib import Path
import os
import json
#from google.colab import drive
#drive.mount('/content/gdrive')
#Add navigation control to zoom map
controls = {"navigation": True, "layer-control": True, "control-grid": True}
m = Map(center=[1.98, 50.61], zoom=11,pitch=60,controls=controls)
m.add_basemap("Esri.WorldImagery")

# Enable 3D terrain with default settings
m.add_3d_terrain()
#move to Mt Etna
m.set_center(14.98956214,37.7518492)
m
```

## Script points to Mt Vesuvius
Configure the view as required

Create a bookmark named Mt Vesuvius

Use bookmarks to flip between views

```python
m.set_center(14.423756,40.8223905)
m.set_zoom(12)
#Again configure the image as you like and create bookmark Mt Vesuvius
m
```

## Prepopulate the bookmarks in line using the add_control_grid function
Exercise the US Bookmarks

Export the bookmarks file

In your downloads directory rename the latest bookmarks.json file to USBookmarks.json

```python
m2 = Map(center=[-0.2664009,51.5287398], zoom=9,controls=controls)
m2.add_basemap("Esri.WorldImagery")

# Enable 3D terrain with default settings
m2.add_3d_terrain()
#move to Mt St Helens
#set_center()
# You have the option of prepopulating the bookmarks tools with inline code
m2.add_control_grid(default_controls=["fullscreen", "north", "terrain", "search", "basemap", "measure", "bookmark", "viewState","vectorDataset",
                    "inspect","minimap","geoEditor"],
                   bookmark_options={
                        "bookmarks": [
                            {
                                "id": "nyc",
                                "name": "New York City",
                                "lng": -74.006,
                                "lat": 40.7128,
                                "zoom": 12,
                                "pitch": 0,
                                "bearing": 0,
                                "createdAt": 0,
                            },
                            {
                                "id": "sf",
                                "name": "San Francisco",
                                "lng": -122.4194,
                                "lat": 37.7749,
                                "zoom": 12,
                                "pitch": 0,
                                "bearing": 0,
                                "createdAt": 0,
                            },
                        ],
},
    rows=4, columns=3,collapsed=False)
m2
```
